## Imports

```python
import sys
print(sys.executable)
import pymupdf
import os
import site
print(site.getusersitepackages())
import cv2
import matplotlib.pyplot as plt
# allows inline plotting
%matplotlib inline

import pytesseract
from pytesseract import Output
# change this path name to where you have stored tesseract.exe
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# regular expression
import re

from pathlib import Path
import shutil
import fitz
from PIL import Image
```

# Support Functions

```python
#reads pdf and converts to an image
def pdf_to_images(pdf_path, output_folder=r"C:\Users\Josh\CV & NLP\Labs\CA2\ca2-assignment-JoshDuffy3\image_output", zoom = 2):
        doc = fitz.open(pdf_path)
        image_paths = []

        for page_number in range(len(doc)):
           page = doc[page_number]
           matrix = fitz.Matrix(zoom, zoom)
           pix = page.get_pixmap(matrix=matrix)

           image_path = output_folder / f"{pdf_path.stem}_page_{page_number + 1}.png"
           pix.save(image_path)

           image_paths.append(image_path)

        return image_paths

#returns all files in a folder     
def get_filepaths(path):
    folder = Path(path)

    files = [file for file in folder.iterdir() if file.is_file()]
    
   

    return files

#extracts text from an image
def extract_text(image_path):
    image_path = Path(image_path)

    image = Image.open(image_path)

    text = pytesseract.image_to_string(image)

    return text
```

# NLP

## Convert PDF to Images and Copy images to output folder

```python
input_folder = r"C:\Users\Josh\CV & NLP\Labs\CA2\ca2-assignment-JoshDuffy3\data"
output_folder = Path(r"C:\Users\Josh\CV & NLP\Labs\CA2\ca2-assignment-JoshDuffy3\image_output")

input_files = get_filepaths(input_folder)
all_image_paths = []

for file in input_files:
    suffix = file.suffix.lower()
    
    if suffix == ".pdf":
        image_paths = pdf_to_images(file, output_folder)
        all_image_paths.extend(image_paths)

    elif suffix in [".png", ".jpg", ".jpeg", ".bmp", ".tfif"]:
        destination = output_folder / file.name
        shutil.copy(file, destination)
        all_image_paths.append(destination)

    else:
        #ignore anything else
        pass
    
for path in all_image_paths:
    print(path)
```

## Extract text from images

```python
image_files = get_filepaths(output_folder)

ocr_results = []

for image_file in image_files:
    print(f"Processing: {image_file.name}")

    extracted_text = extract_text(image_file)

    ocr_results.append({
        "file_name": image_file.name,
        "file_path": str(image_file),
        "extracted_text": extracted_text
    })

print("OCR complete")
```

```python
for result in ocr_results:
    print("=" * 80)
    print(f"File: {result['file_name']}")
    print("-" * 80)
    print(result["extracted_text"])
    print("\n")
```
